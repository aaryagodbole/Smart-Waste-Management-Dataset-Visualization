```python
# 📌 STEP 1: Install Required Libraries (if not installed)
!pip install seaborn scikit-learn pandas matplotlib

# 📌 STEP 2: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from google.colab import files

# 📌 STEP 3: Upload CSV File
print("Please upload the 'Waste_Management_and_Recycling_India.csv' file")
uploaded = files.upload()  # Manually upload the dataset

# 📌 STEP 4: Load the Dataset
df = pd.read_csv("Waste_Management_and_Recycling_India.csv")  # Change filename if needed

# 📌 STEP 5: Display First Few Rows to Verify
print("Dataset Preview:")
print(df.head())

# 📌 STEP 6: Check Column Names
print("\nColumn Names in Dataset:")
print(df.columns)

# 📌 STEP 7: Remove Missing Values (if any)
df = df.dropna(subset=['Waste Generated (Tons/Day)', 'Recycling Rate (%)'])

# 📌 STEP 8: Select Features for Regression
X = df[['Waste Generated (Tons/Day)']].values.reshape(-1, 1)  # Independent variable
y = df['Recycling Rate (%)'].values  # Dependent variable

# 📌 STEP 9: Fit Linear Regression Model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)  # Predicted values

# 📌 STEP 10: Plot Regression Line
plt.figure(figsize=(10, 6))
#sns.scatterplot(x=X.flatten(), y=y, color='blue', alpha=0.6, label='Actual Data')  # Scatter plot
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')  # Regression line
plt.xlabel('Waste Generated (Tons/Day)')
plt.ylabel('Recycling Rate (%)')
plt.title('Linear Regression: Waste Generated vs Recycling Rate')
plt.legend()
plt.show()
```
